```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("IrisSQL").getOrCreate()
```

```python
file_path = "/home/spark/Downloads/archive/Iris.csv"
df = spark.read.csv(file_path, header=True, inferSchema=True)
```

```python
df.createOrReplaceTempView("iris_table")
```

```python
spark.sql("SELECT * FROM iris_table LIMIT 150").show()
```

```python
spark.sql("""
SELECT Species, COUNT(*) as count
FROM iris_table
GROUP BY Species
""").show()
```

```python
spark.sql("""
SELECT Species, AVG(SepalLengthCm) as avg_sepal_length
FROM iris_table
GROUP BY Species
""").show()
```

```python
spark.sql("""
SELECT *
FROM iris_table
WHERE SepalLengthCm > 5
""").show()
```

```python
spark.sql("""
SELECT Species, MAX(SepalLengthCm) as max_sepal
FROM iris_table
GROUP BY Species
""").show()
```

```python
spark.sql("""
SELECT *
FROM iris_table
WHERE PetalLengthCm BETWEEN 3 AND 5
""").show()
```

```python
spark.sql("""
SELECT Species, SepalLengthCm, PetalLengthCm
FROM iris_table
ORDER BY SepalLengthCm DESC
LIMIT 150
""").show()
```

```python
spark.sql("""
SELECT *
FROM iris_table
WHERE SepalLengthCm > 6 AND PetalWidthCm > 1.5
""").show()
```

```python
spark.sql("""
SELECT 
    Species,
    COUNT(*) as toplam_sayi,
    AVG(SepalLengthCm) as ort_sepal,
    AVG(PetalLengthCm) as ort_petal,
    MAX(SepalWidthCm) as maks_sepal_genislik,
    MIN(PetalWidthCm) as min_petal_genislik
FROM iris_table
GROUP BY Species
""").show()
```

```python
spark.sql("""
SELECT *
FROM iris_table
WHERE Species = 'Iris-virginica'
LIMIT 150
""").show()
```

```python
spark.sql("""
SELECT 
    Species,
    COUNT(*) as sayi,
    COUNT(*) * 100.0 / (SELECT COUNT(*) FROM iris_table) as yuzde
FROM iris_table
GROUP BY Species
""").show()
```

```python
spark.sql("""
SELECT *
FROM iris_table
WHERE SepalLengthCm > 7 OR PetalWidthCm > 2
""").show(150)
```

```python
spark.sql("""
SELECT *
FROM iris_table
WHERE SepalLengthCm > (SELECT AVG(SepalLengthCm) FROM iris_table)
""").show()
```

```python
spark.sql("""
SELECT Species, SepalLengthCm, PetalLengthCm
FROM iris_table
ORDER BY Species ASC, SepalLengthCm DESC
LIMIT 150
""").show()
```
